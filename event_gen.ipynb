```python
""" 
#Python script to generate dates --> format of output should be:
{   title: Past or Future event (categoria eventos ou "j치 faltou mais que")
    date_iso: "2024-01-10T18:00:00",
    date: "01 de Outubro de 2024",
    label: "Pedido de casamento 游눐"}
"""

from datetime import datetime, timedelta
from dateutil.relativedelta import relativedelta
import json

event_list = []
due_date = '2034-11-10T00:00:00'
start_date = '2025-12-25T00:00:00' #xmas 2025 

class Event():
    def __init__(self,  date,  label):
        self.date = date
        self.label = label

class PastEvent(Event):
    def write_output(self, due_date):
        #output date is average bt event_date and due_date -> "j치 faltou mais que"
        dic = {}
        diff = datetime.fromisoformat(due_date) - datetime.fromisoformat(self.date)
        avg_date = datetime.fromisoformat(self.date) + diff/2
        dic['title'] = 'PastEvent'
        dic['date_iso'] = avg_date.isoformat() #DATA PARA APARECER A INFO
        data = datetime.fromisoformat(self.date)
        meses = ['janeiro','fevereiro','mar칞o','abril','maio','junho','julho','agosto','setembro','outubro','novembro','dezembro']
        dic['date_pt'] = f'{data.day} de {meses[data.month-1]} de {data.year}' #DATA DO EVENTO NO PASSADO
        dic['label'] = self.label
        return dic

class FutureEvent(Event):
    def write_output(self, due_date):
        #output date is self.date
        dic = {}
        dic['title'] = 'FutureEvent'
        dic['date_iso'] = self.date
        data = datetime.fromisoformat(self.date)
        meses = ['janeiro','fevereiro','mar칞o','abril','maio','junho','julho','agosto','setembro','outubro','novembro','dezembro']
        dic['date_pt'] = f'{data.day} de {meses[data.month-1]} de {data.year}'
        dic['label'] = self.label
        return dic
    

"""Define future events!"""
#Each 10k multiple hours that are left
for i in range(1,9):
    date = datetime.fromisoformat(due_date) - timedelta(hours=10000*i)
    event_list.append( FutureEvent(date.isoformat(), f'J치 s칩 faltam {10*i}.000 horas!') )
#Each 1k multiple hours that are left adter 10k
for i in range(1,11):
    date = datetime.fromisoformat(due_date) - timedelta(hours=1000*i)
    event_list.append( FutureEvent(date.isoformat(), f'J치 s칩 faltam {1*i}.000 horas!') )

#Each 500 multiple days that are left
for i in range(1,8):
    date = datetime.fromisoformat(due_date) - timedelta(days=500*i)
    event_list.append( FutureEvent(date.isoformat(), f'J치 s칩 faltam {500*i} dias!') )
#Each 100 multiple days that are left after 500
for i in range(1,6):
    date = datetime.fromisoformat(due_date) - timedelta(days=100*i)
    event_list.append( FutureEvent(date.isoformat(), f'J치 s칩 faltam {100*i} dias!') )

#Each year
for i in range(1,10):
    date = datetime.fromisoformat(due_date) - relativedelta(years=i)
    if i==1: event_list.append( FutureEvent(date.isoformat(), f'Falta agora {i} ano! PS: Parab칠ns Constan칞a! :P') )
    else: event_list.append( FutureEvent(date.isoformat(), f'Faltam agora {i} anos. PS: Parab칠ns Constan칞a! :P') )


"""Define past events!"""
event_list.append( PastEvent('2016-07-10T22:30:00', f'Portugal ganhou o Euro 2016') )
event_list.append( PastEvent('2016-09-15T00:00:00', f'A Constan칞a e o Jo칚o conheceram-se e o amor nasceu') )
event_list.append( PastEvent('2017-11-01T14:00:00', f'O almo칞o em casa da Constan칞a foi prometido') )
event_list.append( PastEvent('2017-09-11T22:30:00', f'O JECDP nasceu oficialmente') )


"""Write them into file"""
for i,event in enumerate(event_list):
    event_list[i] = event.write_output(due_date)

with open("event_list.json", "w", encoding="utf-8") as f:
    json.dump(event_list, f, indent=4)
```
